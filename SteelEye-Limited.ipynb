```python
import requests
from bs4 import BeautifulSoup as BS

import urllib.request
import zipfile
import os
import csv
import xml.etree.ElementTree as ET
```

```python
url = "https://registers.esma.europa.eu/solr/esma_registers_firds_files/select?q=*&fq=publication_date:%5B2021-01-17T00:00:00Z+TO+2021-01-19T23:59:59Z%5D&wt=xml&indent=true&start=0&rows=100"

response = requests.get(url)
response.encodingn = "utf-8"

XML_Data = BS(response.content, "xml")
XML_Data
```

```python
Urls = XML_Data.find("result", {"name":"response"}).findAll("str",{"name":"download_link"})

for url in Urls:
    print(url.text)
```

```python

# Destination folder to extract the zip files to
dest_folder = r"C:\Users\nanim\Dropbox\Python_Engineer\Extracted_Files"

# Download and extract each zip file
for url in Urls:
    url = url.text
    filename = os.path.basename(url)
    urllib.request.urlretrieve(url, filename)
    with zipfile.ZipFile(filename, "r") as zip_ref:
        zip_ref.extractall(dest_folder)
    os.remove(filename)
```

```python


# Set the path to the folder containing the XML files
folder_path = r"C:\Users\nanim\Dropbox\Python_Engineer\Extracted_Files"

# Set the output CSV file path and open it in write mode
csv_path = "output.csv"
with open(csv_path, "w", newline="") as csv_file:
    # Define the column names for the CSV file
    fieldnames = ["Id", "FullNm", "ClssfctnTp", "CmmdtyDerivInd", "NtnlCcy", "Issr"]
    
    # Create a CSV writer object
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    
    # Write the header row to the CSV file
    csv_writer.writeheader()
    
    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        # Get the full path to the file
        file_path = os.path.join(folder_path, filename)
        
        # Parse the XML file
        tree = ET.parse(file_path)
        root = tree.getroot()
        
        # Loop through all FinInstrmGnlAttrbts elements in the XML file
        for instrm in root.findall(".//{urn:iso:std:iso:20022:tech:xsd:auth.036.001.02}FinInstrmGnlAttrbts"):
            # Extract the data for each column
            row = {
                "Id": instrm.find("{urn:iso:std:iso:20022:tech:xsd:auth.036.001.02}Id").text,
                "FullNm": instrm.find("{urn:iso:std:iso:20022:tech:xsd:auth.036.001.02}FullNm").text,
                "ClssfctnTp": instrm.find("{urn:iso:std:iso:20022:tech:xsd:auth.036.001.02}ClssfctnTp").text,
                "CmmdtyDerivInd": instrm.find("{urn:iso:std:iso:20022:tech:xsd:auth.036.001.02}CmmdtyDerivInd").text,
                "NtnlCcy": instrm.find("{urn:iso:std:iso:20022:tech:xsd:auth.036.001.02}NtnlCcy").text,
#                 "Issr": instrm.find("{urn:iso:std:iso:20022:tech:xsd:auth.036.001.02}Issr").text
            }
            
            # Write the row to the CSV file
#             writer = csv_writer.writerow(row)
#             writer.writerows(data)
            csv_writer.writerow(row)
```
